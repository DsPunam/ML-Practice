```python
import pandas as pd 
df=pd.read_csv("decision_tree.csv")
df
```

```python
inputs=df.drop("Salary more than 100k",axis="columns")
target=df["Salary more than 100k"]
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
company=LabelEncoder()
job=LabelEncoder()
degree=LabelEncoder()
```

```python
inputs["company_c"]=company.fit_transform(inputs["Company"])
inputs["job_c"]=job.fit_transform(inputs["Job"])
inputs["degree_c"]=degree.fit_transform(inputs["Degree"])
```

```python
inputs
```

```python
new_inputs=inputs.drop(["company_c","job_c","degree_c"],axis="columns")
new_inputs
```

```python
from sklearn import tree
```

```python
reg=tree.DecisionTreeClassifier()
reg.fit(new_inputs,target)
```

```python
reg.predict([[2,3,1]])
```
