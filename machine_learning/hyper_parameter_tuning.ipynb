```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["flower"]=iris.target
df
```

```python
df["flower"]=df["flower"].apply(lambda x:iris.target_names[x])
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
model=svm.SVC(kernel="rbf",C=30,gamma="auto")
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
kernel=["rbf","linear"]
C=[1,10,20]
avg_score={}
for kval in kernel:
    for cval in C:
        cross_val=cross_val_score(svm.SVC(kernel=kval,C=cval,gamma="auto"),iris.data,iris.target,cv=5)
        avg_score[kval+"_"+ str(cval)]=np.average(cross_val)
avg_score
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(svm.SVC(gamma="auto"),{
    "C":[1,10,20],
    "kernel":["rbf","linear"]
},cv=5,return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df1=pd.DataFrame(clf.cv_results_)
df1
```

```python
df1[["param_C","param_kernel","mean_test_score"]]
```

```python
from sklearn.model_selection import RandomizedSearchCV
cs=RandomizedSearchCV(svm.SVC(gamma="auto"),{
    "C":[1,10,20],
    "kernel":["rbf","linear"]
},cv=5,return_train_score=False,n_iter=2)
cs.fit(iris.data,iris.target)
pd.DataFrame(cs.cv_results_)[["param_C","param_kernel","mean_test_score"]]
```
