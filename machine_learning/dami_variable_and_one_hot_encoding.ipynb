```python
import pandas as pd 
df=pd.read_csv("town_area_price_file.csv")
df
```

```python
dummies=pd.get_dummies(df.Town) #to create dummy variable coloumn
dummies
```

```python
#to join the two dataset
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(["Town","West Windsor"],axis="columns")
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop(["Price"],axis="columns")
x
```

```python
y=final.Price
y
```

```python
#for training my data
model.fit(x,y)
```

```python
model.predict([[2910,0,1]])
```

```python
model.predict([[5500,1,0]])
```

```python
#to know the accuracy of the model
model.score(x,y)
```
